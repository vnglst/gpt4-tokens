```python
import base64

# Source url: https://github.com/openai/tiktoken/blob/main/tiktoken_ext/openai_public.py
# Url: https://openaipublic.blob.core.windows.net/encodings/cl100k_base.tiktoken
file_name = "cl100k_base.tiktoken"

with open(file_name, 'r') as f:
    lines = f.readlines()

decoded_lines = []
failed = []

for line in lines:
    base64_line = line.strip().split(' ')[0]
    try:
        decoded_line = base64.b64decode(base64_line).decode("utf-8")
        decoded_lines.append(decoded_line)
    except UnicodeDecodeError:
        failed.append(line)
        # print(f"Failed to decode: {base64_line}")

print(f"failed to decode {len(failed)} out of {len(lines)}")
print(f"decoded {len(decoded_lines)} lines ")
# print(f"failed: {failed}")
print(f"decoded:\n{"\n".join(decoded_lines)}")

with open("cl100k_base.txt", 'w') as f:
    for line in decoded_lines:
        f.write(line + '\n')
```
