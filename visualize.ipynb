```python
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3

conn = sqlite3.connect('tokens.db')  
cursor = conn.cursor()

tokens_df = pd.read_sql_query("SELECT type FROM tokens", conn)
token_types = tokens_df['type'].value_counts()
token_types = token_types[:3]

print(token_types)

plt.bar(token_types.index, token_types.values)
plt.xticks(rotation=45)
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Token Types')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3

conn = sqlite3.connect('tokens.db')  
cursor = conn.cursor()

tokens_df = pd.read_sql_query("SELECT lang FROM tokens WHERE type IS 'lang'", conn)
token_langs = tokens_df['lang'].value_counts()
token_langs = token_langs[:10]

plt.bar(token_langs.index, token_langs.values)
plt.xticks(rotation=45)
plt.xlabel('Languages')
plt.ylabel('Count')
plt.title('Natural Language Tokens')
plt.show()
```

```python
import sqlite3


def display_processed_tokens(db_file):
    conn = sqlite3.connect(db_file)
    cursor = conn.cursor()

    cursor.execute("SELECT * FROM tokens WHERE lang = 'java'")
    
    records = cursor.fetchall()

    for record in records:
        print(record)

    print(f"Count: {len(records)}")

    conn.close()


display_processed_tokens('tokens.db')
```
